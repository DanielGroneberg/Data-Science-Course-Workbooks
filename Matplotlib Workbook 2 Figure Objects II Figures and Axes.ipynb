There is a background canvas figure, you add the axes, then you plot on that set of axes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a**4
```

```python
x = np.arange(0, 10)
y = 2*x
```

```python
y
```

Reminder: np.linspace() gives evenly spaced points: np.linspace(min, max, number of points)

and np.arange() gives integer points in a range, not inclusive of the max parameter

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y);
#plt.show()
```

```python
fig2 = plt.figure()

axes = fig2.add_axes([0, 0, 1, 1])
axes.plot(a, b);
#plt.show()
```

Note: Since this is an object-based approach, it's probably best to create a new object such as fig2 if you're going to create a new plot. Simply running 



>axes = fig.add_axes([0, 0, 1, 1])<br>
>axes.plot(a, b);<br>


doesn't do anything, you have to include 

>fig = plt.figure()

and at that point you might as well just make a new object since it's probably more memory efficent than re-assigning to the existing object

The limits are controlled using .set_lim() i.e the domain and range of the x and y axes. 

The axes given do not apply to the inset plot.

You can pretty much just copy and paste these commands to affect the inset plot. I paste it in Notepad and use ctrl+h to find and replace 

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#SMALL AXES
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 25)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

axes.plot(x, y);
```

Just for illustration purposes for how the figure canvas works, this is setting the inset to start at 1,1:

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#SMALL AXES
axes2 = fig.add_axes([1, 1, .25, .25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 25)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

axes.plot(x, y);
```

In this particular instance this is not wildly useful, but I can see how this might be convenient for making side-by-side figures:

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#SMALL AXES
axes2 = fig.add_axes([1.2, 0, 1, 1])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 25)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

axes.plot(x, y);
```
