```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(dpi = 200)

a = np.linspace(0, 10, 11)
b = a**4

axes1 = fig.add_axes([0, 0, 1,1])
axes1.plot(a, b)
```

Adjusting the DPI in order to make the graph appear less blurry ^^

Tweaking the figure size:

```python
fig = plt.figure(figsize = (1, 2), dpi = 200)

axes1 = fig.add_axes([0, 0, 1,1])
axes1.plot(a, b)
```

The arguments are taken as inches

```python
fig = plt.figure(figsize = (10, 3), dpi = 200)

axes1 = fig.add_axes([0, 0, 1,1])
axes1.plot(a, b)
```

Exporting the figure:
First, I want to see if this has to be done in the same cell as the rest of the code

```python
fig.savefig('new_figure.png')
```

It appears to have saved the line but not the axes, which is also what would have happened if I ran fig.savefig() in the same cell as the other code

In order to get the axes in the final image, I have to set the bbox_inches parameter to 'tight':

bbox refers to the bounding box

```python
fig.savefig('new_figure.png', bbox_inches = 'tight')
```

Aparently this syntax is not going to be supported in future versions of python . . .
