In some instances, (machine learning) the index should be a primary key, which means unique for every row: 0,1,2,3,4 . . . there isn't more than one row using 0

In this table, the Payment ID column is also a unique identifier. It can be looked at using 
arr = data["Team"].unique() where Team is the name of the column.

```python
df.set_index("Payment ID")
```

This call causes the rows to be index based on the Payment ID column values.

Payment ID is the name of the index, not a column, so it is not counted as a column.

```python
df.head()
```

The df.set_index("Payment ID") call is not "set in place" i.e. is not permanent. To make it permanent, you have to do:

df = df.set_index("Payment ID")

This call can only be made 1 time. If you try to do it multiple times, you will get an error, because there is no longer a Payment ID column, as it's been set as the index name.

```python
df =df.set_index("Payment ID")
```

```python
df.head()
```

```python
df.reset_index()
```

This unlocks Payment ID as the index name, and places it back in the table as a column.

Now a new column can be set as the index name.
