```python
df.corr()
```

This is a table showing correlation. On the diagonal, the values are 1, i.e. total bill is 100% correlated to total bill, tip is 100% correlated to tip, etc.

The % correlation is comparing one column to another as they increase, i.e the similarity between the two values as they increase.

Tip and total bill are only 68% correlated, i.e. as the size of the total bill increases, tip is only 68% similar to what would be expected, and vice versa.

CC is close to 0 in this correlation table because it is neither correlated nor inversely correlated, it is roughly random
